# **Cleaning**

```python
def clean_columns(df):
    # strip spaces then snake_case
    df = df.copy()
    df.columns = (
        df.columns
        .str.strip()
        .str.replace("\s+", " ", regex=True)
        .str.replace("%", "pct", regex=False)
        .str.replace(".", "", regex=False)
        .str.lower()
        .str.replace(" ", "_")
    )
    return df

df_area = clean_columns(df_area)
df_geo  = clean_columns(df_geo)
```

```python
df_area.head()
```

```python
df_geo.head()
```

```python
#renaming
rename_map = {
    "region": "state",
    "date": "date",
    "frequency": "frequency",
    "estimated_unemployment_rate_(pct)": "unemployment_rate_pct",
    "estimated_employed": "employed",
    "estimated_labour_participation_rate_(pct)": "labour_participation_pct",
    "area": "area",
    "region_1": "region_group",   # in df_geo
    "longitude": "longitude",
    "latitude": "latitude",
}
df_area = df_area.rename(columns=rename_map)
df_geo  = df_geo.rename(columns=rename_map)
```

```python
# Parse date as day-first (e.g., 31-05-2019)
for d in ["date"]:
    if d in df_area.columns:
        df_area[d] = pd.to_datetime(df_area[d], dayfirst=True, errors="coerce")
    if d in df_geo.columns:
        df_geo[d] = pd.to_datetime(df_geo[d], dayfirst=True, errors="coerce")

# Basic sanity
display(df_area.head(3))
display(df_area.dtypes)
display(df_geo.head(3))
display(df_geo.dtypes)
```

```python
def describe_brief(df, name):
    print(f"\n{name}: shape={df.shape}")
    dmin, dmax = None, None
    if "date" in df.columns:
        dmin = df["date"].min()
        dmax = df["date"].max()
    print("  date range:", dmin, "to", dmax)
    if "state" in df.columns:
        print("  #states:", df["state"].nunique())
    if "area" in df.columns:
        print("  areas:", df["area"].dropna().unique()[:10])

describe_brief(df_area, "Area-split (df_area)")
describe_brief(df_geo,  "Geo (df_geo)")
```

3.**National monthly average**

```python
state_month = (
    df_geo
    .groupby(["state", "date"], as_index=False)
    .agg({
        "unemployment_rate_pct": "mean",
        "labour_participation_pct": "mean",
        "employed": "mean",
       "region1": lambda x: x.mode().iat[0] if len(x.dropna()) else np.nan,
        "latitude": "mean",
        "longitude": "mean",
    })
    .sort_values(["state", "date"])
)
state_month.head()
```

```python
state_month_area = (
    df_area
    .groupby(["state", "area", "date"], as_index=False)
    .agg({
        "unemployment_rate_pct": "mean",
        "labour_participation_pct": "mean",
        "employed": "mean",
    })
    .sort_values(["state", "area", "date"])
)
```

```python
india_monthly = (
    state_month.groupby("date", as_index=False)
    .agg({
        "unemployment_rate_pct": "mean",
        "labour_participation_pct": "mean",
        "employed": "sum"
    })
    .rename(columns={
        "unemployment_rate_pct": "india_unemployment_pct",
        "labour_participation_pct": "india_labour_participation_pct",
        "employed": "india_employed_sum"
    })
)
```

# **Visualisation**

```python
import plotly.express as px
import plotly.graph_objects as go
```

**National unemployment Trend**

```python
fig = px.line(
    india_monthly, x="date", y="india_unemployment_pct",
    title="India  Average Unemployment Rate Over Time",
    labels={"date":"Date","india_unemployment_pct":"Unemployment (%)"}
)
fig.update_layout(yaxis_tickformat=".1f")
fig.show()
```

We can see that the highest average is between the months april,may,june. This was the months when the governmebt imposed the lockdown

# **State Trend**

```python
top_states = (
    state_month.groupby("state")["unemployment_rate_pct"].mean()
    .sort_values(ascending=False).head(10).index.tolist()
)
subset = state_month[state_month["state"].isin(top_states)]

fig = px.line(
    subset, x="date", y="unemployment_rate_pct", color="state",
    title="Top 10 States by Avg Unemployment  Monthly Trend",
    labels={"date":"Date","unemployment_rate_pct":"Unemployment (%)","state":"State"}
)
fig.update_layout(legend_title_text="State", yaxis_tickformat=".1f")
fig.show()
```

Puducherry,Jharkhand,Tamil Nadu are in the top 3

# **Rural vs Urban**

```python
recent_cut = state_month_area["date"].max() - pd.offsets.MonthBegin(6)
recent = state_month_area[state_month_area["date"] >= recent_cut]

fig = px.box(
    recent, x="area", y="unemployment_rate_pct", color="area",
    title="Rural vs Urban  Unemployment Distribution (Recent Months)",
    labels={"unemployment_rate_pct":"Unemployment (%)","area":"Area"}
)
fig.show()
```

Unemployement is more focused in the urban areas

# **Labour Participation vs Unemployement**

```python
fig = px.scatter(
    state_month, x="labour_participation_pct", y="unemployment_rate_pct",
    size="employed", color="region1", hover_name="state",
    title="Participation vs Unemployment  State-Month Bubbles",
    labels={
        "labour_participation_pct":"Labour Participation (%)",
        "unemployment_rate_pct":"Unemployment (%)",
        "region1":"Region Group"
    }
)
fig.update_traces(marker=dict(sizemode="area"))
fig.show()
```

# **Geo map**

```python
latest_date = state_month.groupby("state")["date"].max().reset_index().rename(columns={"date":"latest_date"})
latest_points = latest_date.merge(state_month, on=["state"], how="left")
latest_points = latest_points[latest_points["date"] == latest_points["latest_date"]]

fig = px.scatter_geo(
    latest_points,
    lat="latitude", lon="longitude",
    hover_name="state",
    size="unemployment_rate_pct",
    color="unemployment_rate_pct",
    projection="natural earth",
    title=f"Latest State Unemployment  {latest_points['date'].max().date()}",
)
fig.update_geos(fitbounds="locations", showcountries=True)
fig.show()
```

```python
import pandas as pd

# Define lockdown months (JanMar = before, AprJun = after)
state_month["month"] = state_month["date"].dt.month
state_month["year"] = state_month["date"].dt.year

# Focus on 2020 only
df_2020 = state_month[state_month["year"] == 2020]

before_lockdown = df_2020[df_2020["month"].between(1, 3)]
after_lockdown  = df_2020[df_2020["month"].between(4, 6)]

# Group by state and compute mean unemployment
before_avg = before_lockdown.groupby("state")["unemployment_rate_pct"].mean().reset_index()
after_avg  = after_lockdown.groupby("state")["unemployment_rate_pct"].mean().reset_index()

# Merge for comparison
comparison = before_avg.merge(after_avg, on="state", suffixes=("_Before", "_After"))

# Rename columns
comparison.columns = ["State", "Unemployment Rate Before Lockdown", "Unemployment Rate After Lockdown"]

comparison.head()
```

```python
import plotly.express as px

fig = px.bar(
    comparison.melt(id_vars="State",
                    value_vars=["Unemployment Rate Before Lockdown", "Unemployment Rate After Lockdown"],
                    var_name="Period", value_name="Unemployment Rate"),
    x="State", y="Unemployment Rate", color="Period", barmode="group",
    title="Statewise Unemployment: Before vs After COVID Lockdown (2020)"
)
fig.update_layout(xaxis_tickangle=-45, yaxis_tickformat=".1f")
fig.show()
```
